### Read Excel data

```python
voter_data = pd.read_excel('Copy of Data Analyst at RI SOS Interview Assignment_Haixia Liu.xlsx',
                           sheet_name = 'Data - Voter List', engine='openpyxl')
```

```python
voter_his = pd.read_excel('Copy of Data Analyst at RI SOS Interview Assignment_Haixia Liu.xlsx',
                          sheet_name = 'Data - Voter History', engine='openpyxl')
```

### Merge two tables based on 'VOTERID', use 'inner' join, which selects records in both tables. 

```python
merg = pd.merge(voter_data,voter_his,how='inner',on='VOTERID')
```

### Add new columns

#### Task 1: as to whether a voter voted in all three elections in 2020, stored in column 'Vote_All_2020'. 

Yes - Voted in all three elections in 2020  
No - Not voted in all three elections in 2020 

```python
filt_1 = merg['ELECTION2'].notnull() & merg['ELECTION3'].notnull() & merg['ELECTION4'].notnull()
merg['Vote_All_2020'] = np.where(filt_1,'Yes','No')
```

#### Task 2: as to whether a voter voted in at least one of the past four elections, stored in column 'Vote_At_Least_Once'. 

Yes - voted in at least one of the past four elections  
No - not voted in any of the past four elections

```python
filt_2 = merg['ELECTION1'].notnull() | merg['ELECTION2'].notnull() | merg['ELECTION3'].notnull() | merg['ELECTION4'].notnull()
```

```python
merg['Vote_At_Least_Once'] = np.where(filt_2,'Yes','No')
```

#### Task3 : decade voter registered to vote in Rhode Island, stored in column 'Decade_Registered'.

```python
merg['Decade_Registered'] = merg['DATEEFFECTIVE'].apply(lambda x: str(x).split('/')[2][:3]+'0s')
```

### Save final table into Excel file

```python
with pd.ExcelWriter('Copy of Data Analyst at RI SOS Interview Assignment_Haixia Liu.xlsx',
                    mode='a') as writer:  
    merg.to_excel(writer, sheet_name='Tasks 1-3 - Merged Datasets',index=False)
```
