```python
d_parser = lambda x: pd.datetime.strptime(x,'%Y-%m-%d %I-%p')
df = pd.read_csv('ETH_1h.csv', parse_dates=['Date'], date_parser=d_parser)
```

```python
df.head()
```

```python
# only see data in 2020
filt = df['Date']>='2020'
df[filt]
```

```python
# only see data in 2019
filt = (df['Date']>='2019') & (df['Date']<'2020')
df[filt]  
```
